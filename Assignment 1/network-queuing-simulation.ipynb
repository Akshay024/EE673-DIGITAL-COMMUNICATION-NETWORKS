# Network Queuing Simulation

```python
import numpy as np
import matplotlib as mpt
import matplotlib.pyplot as plt
import random
```

**average queue length (in bits) vs p**

```python
def func(packet_sizes , weights):
    # maximum packet arrival rate
    a_max = 1

    # time horizon
    T = 100

    L = random.choices(packet_sizes, weights, k=1) # bits per packet
    R = 5 # transmission rate
    probability_packet_size =np.arange(0,1,0.01)
    # array to store average queue size for each arrival rate 
    avg_queue_size = np.array(0)
    #print(avg_queue_size)
    queue_size_bits = np.zeros(T+1)
    # loop over various arrival rates
    for i in probability_packet_size:
        for t in range (T+1):
            L=random.choices(packet_sizes,weights,k=1)
            if t==0:
                queue_size_bits[t] = max(i*L[0] - 5,0)
            else:
                queue_size_bits[t] = max(queue_size_bits[t-1]+i*L[0]-5,0)
        avg_queue_size = np.append(avg_queue_size,np.mean(queue_size_bits))
        #print(avg_queue_size[i])
    fig, ax = plt.subplots()
    ax.plot(probability_packet_size, avg_queue_size[1:])
    ax.set_title('Network queue simulation')
    ax.set_ylabel('Queue size (bits)')
    ax.set_xlabel('Arrival Rate')
```

```python
# part i
# sampling from a set of values to simulate random packet arrival
packet_sizes = [2,4,6,8]
weights = [0.25,0.25,0.25,0.25]
#weights = [0,0,0,1]
func(packet_sizes,weights)
```

```python
# part ii
# sampling from a set of values to simulate random packet arrival
packet_sizes = [2,4,6,8]
weights = [0,0.5,0.5,0]
#weights = [0,0,0,1]
func(packet_sizes,weights)
```

```python
# part iii
# sampling from a set of values to simulate random packet arrival
packet_sizes = [2,4,6,8]
weights = [0.5,0,0,0.5]
#weights = [0,0,0,1]
func(packet_sizes,weights)
```

```python
# part iv
# sampling from a set of values to simulate random packet arrival
packet_sizes = [2,4,6,8]
weights = [0,0,0,1]
#weights = [0,0,0,1]
func(packet_sizes,weights)
```
